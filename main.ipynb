Sobre Obesidade:

-   Filtre as informações corretas para apresentar os resultados de obesidade.
-   Os dados entre homens mulheres são parecidos? Informe com números.
-   Limpe os dados do DataFrame, criando uma coluna de nome 'Obesity' que conterá os valores de obesidade. Transforme em
    float as colunas que porventura foram importadas como string.

```python
# Filtrando DataFrame retirando o valores sem dados e criando uma cópia
df_obesidade_filtrado = df_obesidade.loc[df_obesidade['Obesity (%)'] != 'No data'].copy()

# Criando nova coluna Com o valores transformados de string para float(porcentagem)
df_obesidade_filtrado['Obesity'] = df_obesidade_filtrado['Obesity (%)'].apply(lambda x: float(x.split(' ')[0]))

# Limpando coluna ['Obesity (%)']
df_obesidade_filtrado = df_obesidade_filtrado.drop(columns=['Obesity (%)'])
df_obesidade_filtrado
```

```python
# Filtrando a média de obesidade por sexo e arredondando
media_obesidade_por_sexo = df_obesidade_filtrado.groupby('Sex')['Obesity'].mean()
print(round(media_obesidade_por_sexo, 2))

# Diferença entre homem e mulher na obesidade
male_obesity = df_obesidade_filtrado.loc[df_obesidade_filtrado['Sex'] == 'Male']['Obesity'].mean()
female_obesity = df_obesidade_filtrado.loc[df_obesidade_filtrado['Sex'] == 'Female']['Obesity'].mean()

# Printando resultados
print(f'Média de obesidade entre homens: {male_obesity:.2f}%')
print(f'Média de obesidade entre mulheres: {female_obesity:.2f}%')
```
