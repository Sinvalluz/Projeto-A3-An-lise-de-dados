```python
import pandas as pd

# Carregar o dataset
# df = pd.read_csv('obesity-cleaned.csv')
df = pd.read_csv('./datasets/obesidade/obesity-cleaned.csv')

# Remover a coluna desnecessria
df = df.drop(columns=['Unnamed: 0'])

# Extrair apenas a parte numrica da coluna 'Obesity (%)'
df['Obesity (%)'] = df['Obesity (%)'].str.extract(r'(\d+\.\d+)').astype(float)

# Selecionar colunas relevantes
df_filtered = df[['Country', 'Year', 'Sex', 'Obesity (%)']]

# Agrupar os dados por pas, ano e sexo para calcular a mdia de obesidade
summary_df = df_filtered.groupby(['Country', 'Year', 'Sex']).mean().reset_index()

import matplotlib.pyplot as plt

# Configure the plot
plt.figure(figsize=(12, 8))

# Filtrar countries Para visualizao (5 countries)
countries = summary_df['Country'].unique()[:5]

for country in countries:
    country_data = summary_df[(summary_df['Country'] == country) & (summary_df['Sex'] == 'Both sexes')]
    plt.plot(country_data['Year'], country_data['Obesity (%)'], label=country)

# Set plot labels and title
plt.xlabel('Ano')
plt.ylabel('Porcentagem de Obesidade')
plt.title('Tendncias de Obesidade por Pas (1975-2016)')
plt.legend(title='Pas')
plt.grid(True)
plt.show()
```

```python

# Carregar o dataset
df = pd.read_csv('obesity-cleaned.csv')

# Remover a coluna desnecessria
df = df.drop(columns=['Unnamed: 0'])

# Extrair apenas a parte numrica da coluna 'Obesity (%)'
df['Obesity (%)'] = df['Obesity (%)'].str.extract(r'(\d+\.\d+)').astype(float)

# Selecionar colunas relevantes
df_filtered = df[['Country', 'Year', 'Sex', 'Obesity (%)']]

# Calcular a mdia de obesidade por sexo
mean_obesity_by_sex = df_filtered.groupby('Sex')['Obesity (%)'].mean().reset_index()

# Exibir os resultados
print(mean_obesity_by_sex)

# Comparar os dados entre homens e mulheres
male_obesity = df_filtered[df_filtered['Sex'] == 'Male']['Obesity (%)'].mean()
female_obesity = df_filtered[df_filtered['Sex'] == 'Female']['Obesity (%)'].mean()

print(f'Mdia de obesidade entre homens: {male_obesity:.2f}%')
print(f'Mdia de obesidade entre mulheres: {female_obesity:.2f}%')
```
